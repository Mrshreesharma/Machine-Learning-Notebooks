```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import matplotlib
matplotlib.rcParams.update({'font.size': 12})
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import seaborn as sns 
```

```python
cancer=load_breast_cancer()
```

```python
print(cancer.feature_names)
```

```python
print(cancer.target_names)
```

```python

cancer_df=pd.DataFrame(np.c_[cancer['data'],cancer['target']],columns = np.append(cancer['feature_names'],['target']))
cancer_df.head()
```

```python
X = cancer_df[cancer_df.columns[:-1]]
y = cancer_df.target
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_predict = svc_model.predict(X_test)
cm = np.array (confusion_matrix(y_test, y_predict, labels=[0,1]))
confusion = pd.DataFrame(cm,index=['+ve Cancer','-ve Cancer'],columns=['Predicted cancer','predicted healthy'])
confusion
```

```python
print(classification_report(y_test,y_predict))
```

What does the recall metric imply?

The recall metric implies to the ability of the model to predict the results which are bieng correctly classified. 

it is ratio of true positives with respect to the sum of true positives and false negatives.
